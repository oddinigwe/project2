```python
#to enable visualizations 
%matplotlib inline

# First, import pandas, a useful data analysis tool especially when working with labeled data
import pandas as pd

# import seaborn, visualization library in python 
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

# Next, we'll load the cereal dataset, which is in the specified directory below
cereal = pd.read_csv("C:\\Users\\oddin\\Desktop\\cereal_project\\cereal.csv")

# Next, display the first 15 rows and all columns of the iris dataframe, good way to see the colum headings for the dataset
cereal.head(15)
```

```python
cereal1 = pd.read_csv("C:\\Users\\oddin\\Desktop\\cereal_project\\cereal.csv",index_col='cereal name')
cereal1
```

```python
cereal2=cereal1[['calories']]
cereal2
```

```python
cereal2.plot(kind='bar',rot=90)
```

```python
# Use the .plot extension from Pandas dataframes to plot things
# Use this to make a scatterplot of the Cereal dataset features-calories on the x-axis and rating on the y-axis.
cereal.plot(kind="scatter", x="calories", y="rating")
```

```python
# A good way to complement the boxplot is by using the Seaborn's striplot
# 
# Use jitter=True so that all the points are not represented(clustered) on the same axis,
#this allows the data to be properly represented
# 
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
# added size to make the change the size of the dots i.e. bigger or smaller
# changed edge color
ax = sns.boxplot(x="type", y="calories", data=cereal)
ax = sns.stripplot(x="type", y="calories", data=cereal, jitter=True, size = 6, edgecolor="black")
```

```python
fig,ax = plt.subplots(figsize=(15,15))
cereal['cereal name'].value_counts(sort=False).plot(kind='pie')
plt.title('Cereal Nutritional Data',fontsize=15)
plt.xlabel('cereal name',fontsize=15)
plt.ylabel('calories',fontsize=15)
```
